```python
bmi_index = float(input("Enter the BMI Index: "))
if (bmi_index < 18.5):
    print("Underweight")
elif (18.5 <= bmi_index <=24.9):
    print("Normal Range")
elif (25.0 <= bmi_index <= 29.9):
    print("Over Weight")
elif (30.0 <= bmi_index <= 34.9):
    print("Very Overweight - Class I")
elif (35.0 <= bmi_index <= 39.9):
    print("Very Overweight - Class II")
else:
    print("Very Overweight - Class III")
```
