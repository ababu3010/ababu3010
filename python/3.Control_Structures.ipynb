```python
age = int(input("Enter the age: "))
```

```python
if (age < 18):
    print("Children")
elif (age < 35):
    print("Adult")
elif (age < 59):
    print("Citizen")
else:
    print("Senior Citizen")
```

```python
lists = [9, 25, 39, 40, 79]
```

```python
for age in lists:
    if (age < 18):
        print("Children")
    elif (age < 35):
        print("Adult")
    elif (age < 59):
        print("Citizen")
    else:
        print("Senior Citizen")
```
